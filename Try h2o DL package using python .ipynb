DEEP LEARNING

```python
# H20 package:

import h2o
h2o.init(nthreads = 2, max_mem_size = 6)
```

```python
# covert data into h20 objects:
data = h2o.H2OFrame(df2)
```

```python
data.names
```

```python
# SPlit data:
y = "BAD"
x = data.names
x.remove("BAD")
train, test = data.split_frame([0.5], seed = 29)
```

```python
# Use h2o model:
m = h2o.estimators.deeplearning.H2ODeepLearningEstimator(seed = 29, 
                                                         reproducible = True )

m.train(x, y, train)

m
```

```python
# How model works on test data:

m.model_performance(test)
```

```python
# Get the prediction using model on test dataset:

p = m.predict(test)
```

```python
# See the result:
result = p.as_data_frame()
result.head()
```

```python
# Compare the prediction and real dataset:
compared = p["predict"].cbind(test["BAD"]).as_data_frame()

compared.head()
```

```python
# Accuracy of prediction:

(p["predict"] == test["BAD"]).mean()
```
